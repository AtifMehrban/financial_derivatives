```python
import pandas as pd
import numpy as np
import datetime
pd.set_option('precision' , 2)
```

```python
combined  = pd.concat([msft , aapl])
combined
```

```python
combined1 = pd.concat([msft , aapl]).sort_index()
combined1
```

```python
s4g = combined1[['Symbol' , 'Volume']]
s4g
```

```python
s4g = s4g.reset_index()
s4g
```

```python
pd.DatetimeIndex(s4g['Date'])
```

```python
pd.DatetimeIndex(s4g['Date']).year
```

```python
pd.DatetimeIndex(s4g['Date']).month
```

```python
s4g.insert(2 , 'month' , pd.DatetimeIndex(s4g['Date']).month)
```

```python
s4g
```

```python
grouped = s4g.groupby('Symbol')
```

```python
grouped.ngroups    #no of unique groups
```

```python
grouped.groups
```

```python
def print_groups(groupObject):
    for i , j in groupObject:
        print(i)
        print(j)
```

```python
print_groups(grouped)
```

```python
grouped.size()
```

```python
grouped.get_group('AAPL')
```
